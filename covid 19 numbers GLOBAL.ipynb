```python
import pandas as pd
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', 14)

import matplotlib.pyplot as plt


import numpy as np
from numpy import log as ln
```

```python
input_dir = '/home/cjd/COVID19/'
output_dir = '/home/cjd/COVID19/Output/'
```

```python
cases = pd.read_csv(input_dir + 'time_series_covid19_confirmed_global.csv')
cases = cases.drop(columns=['Lat', 'Long'])
cases = pd.concat([cases.groupby('Country/Region').sum()])
cases = cases.sort_values(by=[cases.columns[-1]], ascending=False)
cases
```

```python
deaths = pd.read_csv(input_dir + 'time_series_covid19_deaths_global.csv')
deaths = deaths.drop(columns=['Lat', 'Long'])
deaths = pd.concat([deaths.groupby('Country/Region').sum()])
deaths = deaths.sort_values(by=[deaths.columns[-1]], ascending=False)
deaths
```

```python
deaths_per_day = deaths.diff(axis='columns')
cases_per_day = cases.diff(axis='columns')
```

```python
top_ten = cases.index[0:10].tolist()
calc = pd.DataFrame()
calc['Cases'] = cases.loc[top_ten,cases.columns[-1]]
calc['Deaths'] = deaths.loc[top_ten,deaths.columns[-1]]
calc['Case Fatality Rate'] =  (calc['Deaths'] / calc['Cases']) * 100.0
calc = calc.sort_values(by=['Case Fatality Rate'], ascending=False)

# calc.style.bar(subset=['Cases'], align='left', color=['orange']).bar(subset=['Deaths'], align='left', color=['red']).format("{: .1f}")
calc.style.bar(subset=['Cases'], align='left', color=['moccasin']).bar(subset=['Deaths'], align='left', color=['lightcoral']).format({"Case Fatality Rate": "{:5,.1f} %",}).set_caption("Ten Countries with Highest Number of Cases Sorted by CFR")
```

```python
cases_doubling_time = ln(2)/ln(1 + cases.loc[top_ten,:].pct_change(axis='columns'))
cases_doubling_time = cases_doubling_time.loc[ :, cases.columns[-14:] ]

cases_doubling_time = cases_doubling_time.sort_values(by=[cases_doubling_time.columns[-1]])

cases_doubling_time.style.background_gradient(cmap='Greens', vmin=1.0, vmax=14.0).format("{:.1f}").set_caption("Doubling Times for Cases in [days] Sorted by Most Recent")
```

```python
deaths_doubling_time = ln(2)/ln(1 + deaths.loc[top_ten,:].pct_change(axis='columns'))
deaths_doubling_time = deaths_doubling_time.loc[ :, deaths.columns[-14:] ]

deaths_doubling_time = deaths_doubling_time.sort_values(by=[deaths_doubling_time.columns[-1]])

deaths_doubling_time.style.background_gradient(cmap='Reds_r', vmin=1.0, vmax=7.0).format("{:.1f}").set_caption("Doubling Times for Deaths in [days] Sorted by Most Recent")
```

```python
pop_dict = [
    {'Country/Region': 'US', 'population': 327.2e6}, 
    {'Country/Region': 'China', 'population': 1.386e9},
    {'Country/Region': 'Italy', 'population': 60.48e6},
    {'Country/Region': 'Iran', 'population': 81.16e6},
    {'Country/Region': 'Spain', 'population': 46.66e6},
    {'Country/Region': 'United Kingdom', 'population': 66.44e6},
    {'Country/Region': 'France', 'population': 66.99e6},
    {'Country/Region': 'Switzerland', 'population': 8.57e6},
    {'Country/Region': 'Korea, South', 'population': 51.47e6},
    {'Country/Region': 'Germany', 'population': 82.79e6},
]

# res = [ sub['Country/Region'] for sub in pop_dict ]


pop = pd.DataFrame(pop_dict)

country_list = pop['Country/Region'].tolist()

pop['per_thou'] = pop['population']/1e3

pop = pop.set_index('Country/Region')

cases_pm = cases.loc[country_list,:].div(pop['per_thou'], axis=0).round(3)

# cases_pm
```

```python
%matplotlib

png_w=10
png_h=10

plt.figure(1, figsize=(png_w, png_h))

country_list = [
'US', 
'Korea, South',
'Italy',
'Spain',
'France',
'United Kingdom',   
'Germany',
# 'China',
# 'Iran',
]

styles = ['b-','g-.','k--','r:','m.-','co-','ys-']

n_days=30

for i,name in enumerate(country_list):

    ax = cases_pm.loc[ name, cases_pm.columns[-n_days:] ].plot(style=styles[i])
    
ax.set_xticks(np.arange(n_days))
temp = cases_pm.columns[-n_days:].to_list()
ax.set_xticklabels(temp, rotation=45)

    
ax.set_title('Cases in Select Countries', fontsize=26)
ax.set_xlabel('Date', fontsize=20)
ax.set_ylabel('Number of Cases Per 1000 Residents', fontsize=20)

ax.legend(country_list, loc='upper left', fontsize=20)

ax.grid(True, alpha=0.25)



plt.savefig(output_dir + 'global_plot_cases.png')
# plt.show()
```
