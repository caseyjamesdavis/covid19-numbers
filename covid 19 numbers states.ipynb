```python
cases_reg = pd.read_csv('/home/cjd/COVID19/us-states.csv')

cases_reg
```

```python
recent = cases_reg.groupby('date')

# recent = recent.get_group('2020-03-29')
recent = recent.get_group(cases_reg['date'].iloc[-1])

recent = recent[recent['deaths']>10]

recent = recent.sort_values(by=['deaths'], ascending=False)

recent['label'] = recent['state'] + '\n' + recent['deaths'].astype(str)

recent
```

```python
%matplotlib
plt.set_cmap('terrain')

# width = 2.0
# height = 1.0

plt.figure(1)
# plt.figure(num=1, figsize=(width, height))

values = recent['deaths'].to_list()
names = recent['label'].to_list()


# values = squarify.normalize_sizes(values, width, height)


ax = squarify.plot(label=names,sizes=values, alpha=.6, text_kwargs={'fontsize':8})

ax.axis('off')
ax.set_title('Deaths as of '+ recent['date'].iloc[0] +' (for states with deaths > 10)',fontsize=23,fontweight="bold")

plt.show()
```

```python
state_data = [
#     {'Province/State': 'New York', 'population': 19.54e6}, 
#     {'Province/State': 'Massachusetts', 'population': 6.902e6}, 
    {'Province/State': 'Vermont', 'population': 626299},
    {'Province/State': 'New Hampshire', 'population': 1.356e6},
    {'Province/State': 'Maryland', 'population': 6.043e6}, 
#     {'Province/State': 'Virginia', 'population': 8.518e6}, 
#     {'Province/State': 'West Virginia', 'population': 1.806e6}, 
#     {'Province/State': 'North Carolina', 'population': 10.38e6},
#     {'Province/State': 'South Carolina', 'population': 5.084e6},
    {'Province/State': 'Florida', 'population': 21.3e6},
]

test = pd.DataFrame(state_data)
test
```

```python
# %matplotlib


plt.figure(2)
ax = plt.gca()



state_list = test['Province/State'].to_list()



styles = ['b-','g-.','k--','r:','m.-','co-','ys-']



for i,name in enumerate(state_list):

    data = states.get_group(name)
    
    data['date'] = pd.to_datetime(data['date'])
    
    ax = data.plot(x='date', y='cases', ax=ax, style=styles[i])
    
# n_days=30
# ax.set_xticks(np.arange(n_days))
# temp = cases_pm.columns[-n_days:].to_list()
# ax.set_xticklabels(temp, rotation=45)

    
ax.set_title('Cases in Select States', fontsize=26)
ax.set_xlabel('Date', fontsize=20)
ax.set_ylabel('Number of Cases', fontsize=20)

ax.legend(state_list, loc='upper left', fontsize=20)

ax.grid(True, alpha=0.25)




plt.show()
```
