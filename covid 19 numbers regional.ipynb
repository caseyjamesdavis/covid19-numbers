```python
import pandas as pd
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', 10)

import matplotlib.pyplot as plt

from numpy import log as ln
```

```python
region = [
'Vermont', 
'New York',
'New Hampshire',
'Massachusetts',
'Quebec',
]
```

```python
cases_reg = pd.read_csv('/home/cjd/COVID19/time_series_19-covid-Confirmed.csv')

cases_reg = cases_reg.drop(columns=['Lat', 'Long', 'Country/Region'])

cases_reg = cases_reg.loc[cases_reg['Province/State'].isin(region)]
cases_reg = cases_reg.set_index('Province/State')

cases_reg = cases_reg.sort_values(by=[cases_reg.columns[-1]], ascending=False)

cases_reg
```

```python
deaths_reg = pd.read_csv('/home/cjd/COVID19/time_series_19-covid-Deaths.csv')

deaths_reg = deaths_reg.drop(columns=['Lat', 'Long', 'Country/Region'])

deaths_reg = deaths_reg.loc[deaths_reg['Province/State'].isin(region)]
deaths_reg = deaths_reg.set_index('Province/State')

deaths_reg = deaths_reg.sort_values(by=[deaths_reg.columns[-1]], ascending=False)

deaths_reg
```

```python
calc_reg = pd.DataFrame()
calc_reg['Cases'] = cases_reg.loc[:,cases_reg.columns[-1]]
calc_reg['Deaths'] = deaths_reg.loc[:,deaths_reg.columns[-1]]
calc_reg['Case Fatality Rate'] =  (calc_reg['Deaths'] / calc_reg['Cases']) * 100.0
calc_reg = calc_reg.sort_values(by=['Case Fatality Rate'], ascending=False).round(1)
calc_reg
```

```python
cases_doubling_time_reg = ln(2)/ln(1 + cases_reg.pct_change(axis='columns'))
cases_doubling_time_reg = cases_doubling_time_reg.loc[ :, cases_reg.columns[-10:] ]

cases_doubling_time_reg.sort_values(by=[cases_doubling_time_reg.columns[-1]]).round(1)
```

```python
deaths_doubling_time_reg = ln(2)/ln(1 + deaths_reg.pct_change(axis='columns'))
deaths_doubling_time_reg = deaths_doubling_time_reg.loc[ :, deaths_reg.columns[-10:] ]

deaths_doubling_time_reg.sort_values(by=[deaths_doubling_time_reg.columns[-1]]).round(1)

deaths_doubling_time_reg

# deaths_doubling_time_reg.style.background_gradient(cmap='Reds_r', vmin=1.0, vmax=10).format("{:.1f}")
```

```python
pop_list = [
    {'Province/State': 'New York', 'population': 19.54e6}, 
    {'Province/State': 'Massachusetts', 'population': 6.902e6}, 
    {'Province/State': 'Quebec', 'population': 8.485e6}, 
    {'Province/State': 'Vermont', 'population': 626299},
    {'Province/State': 'New Hampshire', 'population': 1.356e6},
]

pop = pd.DataFrame(pop_list)
pop = pop.set_index('Province/State')
pop['per_million'] = pop['population']/1e6

# pop

cases_reg_pm = cases_reg.div(pop['per_million'], axis=0).round(1)

cases_reg_pm

```

```python
%matplotlib



styles = ['b-','g-.','k--','r:','m.-','co-','ys-']

for i,name in enumerate(region):

    ax = cases_reg_pm.loc[ name, cases_reg_pm.columns[-10:] ].plot(style=styles[i])


ax.set_title('Cases in Vermont and Surrounding States', fontsize=26)
ax.set_xlabel('Date', fontsize=20)
ax.set_ylabel('Number of Cases Per Million Residents', fontsize=20)

ax.legend(region, loc='upper left', fontsize=20)

ax.grid(True, alpha=0.25)

plt.show()
```
