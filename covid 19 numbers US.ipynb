```python
cases = pd.read_csv(input_dir + 'time_series_covid19_confirmed_US.csv')
cases = cases.drop(columns=['Lat', 'Long_', 'UID', 'code3', 'FIPS'])

cases = pd.concat([cases.groupby('Province_State').sum()])
cases = cases.sort_values(by=[cases.columns[-1]], ascending=False)

cases = cases.drop(index=['Guam', 'Diamond Princess', 'Virgin Islands', 'Northern Mariana Islands', 'American Samoa', 'Grand Princess', 'Puerto Rico'])

cases
```

```python
deaths = pd.read_csv(input_dir + 'time_series_covid19_deaths_US.csv')
deaths = deaths.drop(columns=['Lat', 'Long_', 'UID', 'code3', 'FIPS'])

deaths = pd.concat([deaths.groupby('Province_State').sum()])
deaths = deaths.sort_values(by=[deaths.columns[-1]], ascending=False)

deaths = deaths.drop(index=['Guam', 'Diamond Princess', 'Virgin Islands', 'Northern Mariana Islands', 'American Samoa', 'Grand Princess', 'Puerto Rico'])

pop = deaths[['Population',]].copy()

deaths = deaths.drop(columns=['Population'])


deaths
# pop
```

```python
deaths_per_day = deaths.diff(axis='columns')
cases_per_day = cases.diff(axis='columns')
```

```python
# top_ten = cases.index[0:10].tolist()
calc = pd.DataFrame()

calc['Cases'] = cases.loc[:,cases.columns[-1]]
calc['Deaths'] = deaths.loc[:,deaths.columns[-1]]

calc['Case Fatality Rate'] =  (calc['Deaths'] / calc['Cases']) * 100.0
calc = calc.sort_values(by=['Case Fatality Rate'], ascending=False)


calc.style.bar(subset=['Cases'], align='left', color=['moccasin']).bar(subset=['Deaths'], align='left', color=['lightcoral']).format({"Case Fatality Rate": "{:5,.1f} %",}).set_caption("All States Sorted by CFR")
```

```python
cases_doubling_time = ln(2)/ln(1 + cases.pct_change(axis='columns'))
cases_doubling_time = cases_doubling_time.loc[ :, cases.columns[-14:] ]

cases_doubling_time = cases_doubling_time.sort_values(by=[cases_doubling_time.columns[-1]])

# cases_doubling_time.style.background_gradient(cmap='Greens', vmin=1*7, vmax=4*7).format("{:.1f}").set_caption("Doubling Times for Cases in [days] Sorted by Most Recent")
cases_doubling_time.style.background_gradient(cmap='RdYlGn', vmin=365.25/12.0, vmax=365.25/2.0).format("{:.1f}").set_caption("Doubling Times for Cases in [days] Sorted by Most Recent")
```

```python
deaths_doubling_time = ln(2)/ln(1 + deaths.pct_change(axis='columns'))
deaths_doubling_time = deaths_doubling_time.loc[ :, deaths.columns[-14:] ]

deaths_doubling_time = deaths_doubling_time.sort_values(by=[deaths_doubling_time.columns[-1]])

deaths_doubling_time.style.background_gradient(cmap='RdYlGn', vmin=365.25/12.0, vmax=365.25/2.0).format("{:.1f}").set_caption("Doubling Times for Deaths in [days] Sorted by Most Recent")
```

```python
pop['per_hundred'] = pop['Population']/1e2
pop['per_thou'] = pop['Population']/1e3
pop['per_million'] = pop['Population']/1e6



cases_pc = cases.div(pop['per_hundred'], axis=0).round(3)

cases_pc = cases_pc.loc[ :, cases_pc.columns[-14:] ]
cases_pc = cases_pc.sort_values(by=[cases_pc.columns[-1]], ascending=False)

cases_pc.style.background_gradient(cmap='Blues', vmin=cases_pc.min().min(), vmax=cases_pc.max().max()).format("{:.1f}").set_caption("Number of Cases per 100 Sorted by Most Recent")
```

```python

recent = deaths.loc[ :, deaths.columns[-1:]]

recent = recent.rename(columns={ recent.columns[0]: "deaths" })

recent = recent[recent['deaths']>1000]

recent = recent.reset_index()

recent['label'] = recent['Province_State'] + '\n' + recent[recent.columns[-1]].astype(str)

# recent
```

```python
%matplotlib



png_w=10
png_h=10

plt.figure(1, figsize=(png_w, png_h))


plt.set_cmap('terrain')


values = recent['deaths'].to_list()
names = recent['label'].to_list()


# values = squarify.normalize_sizes(values, width, height)


ax = squarify.plot(label=names,sizes=values, alpha=.6, text_kwargs={'fontsize':10})

ax.axis('off')
ax.set_title('Deaths as of '+ deaths.columns[-1] +' (for states with deaths > 1000)',fontsize=23,fontweight="bold")


plt.subplots_adjust(left=0, bottom=0, right=1, top=0.95, wspace=0, hspace=0)


plt.savefig(output_dir + 'us_plot_treemap_deaths.png')
# plt.show()
```

```python
# %matplotlib

png_w=10
png_h=10

plt.figure(2, figsize=(png_w, png_h))

state_list = [
    'Vermont',
    'New Hampshire',
    'New York',
    'Maryland',
    'Florida',
    'Massachusetts',
    #     'Connecticut',
]

styles = ['b.-','g-.','k--','r:','m.-','co-','ys-']



for i,name in enumerate(state_list):
    
    state_pop = pop['per_thou'].loc[name]

    temp = pd.DataFrame(cases.loc[name, :])

    temp['per capita'] = temp[name] / state_pop

    temp = temp.loc[temp['per capita']>0.1, : ]

    temp = temp.reset_index()



    ax = temp['per capita'].plot(style=styles[i])
    
ax.set_yscale('log')
    
# ax.set_xticks(np.arange(n_days))
# temp = cases.columns[-n_days:].to_list()
# ax.set_xticklabels(temp, rotation=45)

    
ax.set_title('Cases in Select States', fontsize=26)
ax.set_xlabel('Number of Days Since 0.1 Cases per 1000 Residents', fontsize=20)
ax.set_ylabel('Number of Cases per 1000 Residents', fontsize=20)

ax.legend(state_list, loc='upper left', fontsize=20)

ax.grid(True, alpha=0.25)


plt.savefig(output_dir + 'us_plot_cases_log.png')
# plt.show()
```

```python
# %matplotlib

png_w=10
png_h=10

plt.figure(3, figsize=(png_w, png_h))

state_list = [
    'Vermont',
#     'New Hampshire',
#     'New York',
    'Maryland',
    'Florida',
#     'Massachusetts',
#         'Connecticut',
]

styles = ['b.-','g-.','k--','r:','m.-','co-','ys-']

n_days=30

for i,name in enumerate(state_list):

    ax = cases_per_day.loc[ name, cases.columns[-n_days:] ].plot(style=styles[i])
    
ax.set_xticks(np.arange(n_days))
temp = cases.columns[-n_days:].to_list()
ax.set_xticklabels(temp, rotation=45)

    
ax.set_title('Cases per Day in Select States', fontsize=26)
ax.set_xlabel('Date', fontsize=20)
ax.set_ylabel('Number of Cases per Day', fontsize=20)

ax.legend(state_list, loc='upper left', fontsize=20)

ax.grid(True, alpha=0.25)


plt.savefig(output_dir + 'us_plot_cases_per_day.png')
# plt.show()
```
